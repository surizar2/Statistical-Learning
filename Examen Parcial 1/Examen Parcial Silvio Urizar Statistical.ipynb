4. Describa brevemente como se realiza un modelo de predicción para ML

* Para realiza un modelo predictivo de ML, pueden seguirse los siguientes pasos:

6. ¿Cómo se calcula los coeficientes en una regresión lineal?

Una ecuación de la recta viene dada por:

$ y = \theta_0 + \theta_1 x$

Donde:

$\theta_1 = \frac{\Sigma_{i = 1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\Sigma_{i = 1}^{n}(x_i - {x})^2}$

$\theta_0 = \bar{y} - \theta_1\bar{x}$

En estas ecuaciones, se ingresan los valores correspondientes al set de entrenamiento. Deben seguirse los siguientes pasos:
* Obtener el valor promedio de los datos correspondientes a la variable independiente
* Obtener el valor promedio de los datos correspondientes a la variable dependiente
* Utilizar la ecuación de $\theta_1$ con los valores obtenidos
* Con el valor de $\theta_1$ obtenido, se obtiene $\theta_0$

7. ¿Qué es "ACCURACY" en un modelo de ML?

El Accuracy de un modelo es la precisión de este. Este parámetro es utilizado para determinar cuán buen modelo es el generado, y nos indica la cantidad de predicciones correctas de entre todas las predicciones hechas por el modelo generado.

8. ¿Cómo funciona K-FOLD CROSS VALIDATION?

* Para el método K-Fold Cross Validation, se realiza un muestreo n veces para crear n sets de entrenamiento y n sets de pruebas. Con cada set de entrenamiento generado, se entrena el modelo deseado y se obtienen los coeficientes del mismo, y se prueban con el set de pruebas correspondiente para ver la calidad del modelo obtenido. Mediante esto, se generan n modelos, y el modelo definitivo a utilizar es el promedio de los modelos generados.

11. Usted esta interpretando el resultado de un algoritmo de ML que categoriza pacientes de enfermedades terminales, la matriz de confusión devolvio el valor TP con 8 y FN con 15, ¿qué podría concluir del modelo?

Con esta información, se puede obtener el recall del modelo

Recall =$ \frac{\text{TP}}{\text{TP} + \text{FN}} = \frac{8}{23} = 0.35$

Este índice nos dice que del todos los casos categorizados como verdaderos, únicamente el 35% fue catalogado correctamente. Si el modelo que tiene un recall de 35%, y ese desea diagnosticar correctamente casos positivos, este se considera un mal modelo.

12. Usted esta interpretando un modelo de ML que tiene un F1-SCORE de 1, ¿qué podria concluir del modelo?

El F1 score de 1 implica que el modelo obtenido es un modelo perfecto, que puede predecir cualquier valor sin error alguno dada una combinación de variables dependientes.
Esto es muy poco probable, por lo que podría considerarse que el modelo está sobreajustado.

14. Explique los indicadores obtenidos de la matriz de confusión.

La matriz de confusión consta de 4 elementos

* TP: True positive, elementos que son verdaderos, y fueron catalogados con un modelo como positivos.
* FP: False positives, elementos que son falsos, pero fueron catalogados con un modelo como positivos.
* FN: False negatives, elementos que en verdad son positivos, pero el modelo los catalogó como negativos.
* TN: True Negatives, elementos que son negativos, y fueron catalogados con un modelo correctamente como negativos.

El problema que se tiene para resolver es un problema de clasificación, por lo que se opta por realizar el algoritmo de decision Tree

### Acondicionamiento de los datos

```python
from sklearn import preprocessing
```

```python
Data.head()
```

```python
DataFormat = Data.loc[:, Data.columns != 'class'].copy()
Y = Data['class']
LabelEncoder = preprocessing.LabelEncoder()
for value in DataFormat:
    if DataFormat[value].dtype == 'object':
        DataFormat[value] = LabelEncoder.fit_transform(DataFormat[value])

DataFormat.head()
        
```

```python
Data.head()
```

### Set de entrenamiento y de pruebas

```python
from sklearn.model_selection import train_test_split
```

```python
XTrain, XTeset, YTrain, YTest = train_test_split(DataFormat,Y,test_size = 0.8, random_state = 3)
```

### Generando árbol de decisiones

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Modelo = DecisionTreeClassifier(criterion = 'entropy', max_depth= 10, min_samples_split=20)

Modelo.fit(XTrain, YTrain)
```

```python
Predicciones = Modelo.predict(XTeset)
```

```python
Predicciones[0:5]
```

```python
YTest[0:5]
```

### Calificando el modelo

```python
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
Acc = metrics.accuracy_score(YTest,Predicciones)
print(f'Precisión del modelo = {Acc}')
```

```python
Acc2 = metrics.classification_report(YTest,Predicciones)
print(Acc2)
```

### Cuál es la variable más significativa?
